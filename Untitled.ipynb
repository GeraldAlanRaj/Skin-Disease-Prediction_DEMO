```python
train = ImageDataGenerator(rescale = 1/255)
validation = ImageDataGenerator(rescale = 1/255)
```

```python
train_dataset = train.flow_from_directory("/Users/geraldalanraja/Desktop/Skin Prediction/Aggregated Data/Training",
                                          target_size = (200,200),
                                          batch_size = 3,
                                          class_mode = "categorical")
    
```

```python
train_dataset.class_indices
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation="relu", input_shape=(200, 200, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    
    tf.keras.layers.Conv2D(64, (3, 3), activation="relu"),
    tf.keras.layers.MaxPooling2D(2, 2),
    
    tf.keras.layers.Conv2D(128, (3, 3), activation="relu"),
    tf.keras.layers.MaxPooling2D(2, 2),
    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation="relu"),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(3, activation="softmax")
])
```

```python
model.compile(optimizer="adam",
              loss="categorical_crossentropy",
              metrics=["accuracy"])
```

```python
model_fit = model.fit(train_dataset,
                      epochs=10)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image

# Load image
img_path = "/Users/geraldalanraja/Desktop/Skin Prediction/Aggregated Data/Testing/Dysplastic Nevus/IMD436.jpg"
img = image.load_img(img_path, target_size=(200, 200))  # Ensure same size as CNN input

# Display image
plt.imshow(img)
plt.axis("off")
plt.title("Input Image")
plt.show()

# Convert to array and reshape for model input
X = image.img_to_array(img)
X = np.expand_dims(X, axis=0)  # Expand dimensions to match model input shape
X = X / 255.0  # Normalize if trained with normalized data

# Predict
val = model.predict(X)  # Get prediction probabilities
predicted_class = np.argmax(val)  # Get index of highest probability

# Class mapping
class_labels = ["Basal Cell Carcinoma", "Dysplastic Nevus", "Actinic Keratosis"]

# Output result
print(f"Predicted Class: {class_labels[predicted_class]} (Confidence: {val[0][predicted_class]:.2f})")
```
